**Çalışmaların Özeti :**

Simulasyon sonucu oluşturulan veri setinde 7 adet yatak sıcaklığı ve ana motor sıcaklığı yer almaktadır.Bunların yanı sıra akım ve motor devir değerleri de mevcuttur.

Buradaki mevcut çalışmada 3.yatak sıcaklığı kullanılmıştır.

**172800 saniye** yani **48 saatlik** **3. yatak sıcaklık** değerlerinden oluşan veri seti üzerinde trend tespit çalışması gerçekleştirilmiştir. 


Her bir **24 saat içinde 1 adet** olmak üzere **35-42** dereceleri arasında kritik artışların yaşandığı durumlar söz konusudur.

Bu trendlerin tespit edilmesi için hareketli ortalama ve belirli aralıklarla eğim değerlerinin ölçülmesi yöntemlerinden faydalanılmıştır.
- 1800 saniyelik yani 30 dakikalık hareketli ortalama kullanılmıştır.
- Hareketli ortalama sonrasında eğimlerin hesaplanması için kullanılan parametreler aşağıdaki gibidir:
    - trend_vol_threshold : Eğimin kritik trend olarak değerlendirilmesi için gereken eşik değeri
    - window_size = Eğimi hesaplanacak zaman aralağı
    - step_size = Zaman aralıklarının kaç birim kaydırılacağı


- Yapılan bu çalışmada trend_vol_threshold = 60, window_size = 4, step_size = 1 olarak seçilmiştir.

**Gerekli Kütüphaneler**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,8)
import datetime
import math
```

#### Veri Setinin Okunması

```python
df = pd.read_csv("temp_data.csv")
```

```python
plt.plot(df)
plt.title("2 Günlük Press Makinesi Sıcaklık Değerleri");
```

###  Gerekli Fonksiyonların Oluşturulması <a id =2></a>

```python
def preprocess(df,time_range):
    date_index = pd.date_range(start="2021-08-04", periods=172800, freq="1s")
    df.set_index(date_index,inplace=True)
    df.columns = ["Temp"]
    window_count = df.shape[0] // time_range
    window_list = list(range(1,window_count+1))
    return df, window_count, window_list
```

```python
def calc_mov_avg(df,feature,window_list):
    rolmean = df[feature].rolling(1800).mean()
    rolmean_hour = rolmean.resample("30min").mean()
    rolmean_hour.index = window_list
    return rolmean_hour
```

```python
def detect_trend(data, trend_vol_threshold, 
                 window_size, step_size):
    df_trends = pd.DataFrame(columns = ["Date","Time","Trend Degree","IsTrend"])
    for i in range(1, len(data) - window_size + 1,step_size):
        indexes = list(data.loc[i:i + window_size].index)
        coeffs = np.polyfit(indexes, list(data.loc[i:i + window_size].values), 1)
        slope = round(coeffs[-2], 4)
        degree = round(math.degrees(math.atan(slope)),5)
        hour = (i*30)//60
        minute = (i*30)%60
        time = str((i*30)//60) + "." +str((i*30)%60)
        today = datetime.datetime.now()
        day = today.strftime("%d/%m/%Y")
        df_trends.loc[i,"Date"] = day
        if hour > 24:
            day = datetime.datetime.now() + datetime.timedelta(days=1)
            day = day.strftime("%d/%m/%Y")
        df_trends.loc[i,"Date"] = day   
        df_trends.loc[i,"Time"] = time
        df_trends.loc[i,"Trend Degree"] = degree
        if degree > trend_vol_threshold:
            df_trends.loc[i,"IsTrend"] = True
            print("!!! {} tarihinde saat {} 'da/de derecesi {} olan trend tespit edilmiştir !!!".format(day,time,degree))
        elif degree < trend_vol_threshold:
            df_trends.loc[i,"IsTrend"] = False
            print("Tarih : {} Saat : {} Degree : {} Herhangi bir trend tespit edilememiştir. ".format(day,time,degree))
    return df_trends
```

### Trendlerin Tespit Edilmesi <a id =3></a>

```python
df, window_count, window_list = preprocess(df=df, time_range=1800)
```

```python
rolmean_hour = calc_mov_avg(df,"Temp",window_list)
```

```python
df_degree = detect_trend(data=rolmean_hour,trend_vol_threshold=60, window_size=4, step_size=1)
```

```python
df_degree
```

### Yakalanan Trendlerin Gösterimi <a id =4></a>

```python
trend_hour_plot = list()
plt.plot(rolmean_hour,marker="o")
for i in df_degree.index:
    if df_degree.loc[i,"IsTrend"] == True:
        value = rolmean_hour[i]
        plt.annotate(
        'Alert', xy=(i, value),xytext=(i-1, value-2) ,fontsize=13,rotation=90,
            color = "r",arrowprops=dict(arrowstyle='->',lw=2,color="r"))
plt.xticks(np.arange(0,96,2))
plt.title("Yakalanan Trendler");
```
